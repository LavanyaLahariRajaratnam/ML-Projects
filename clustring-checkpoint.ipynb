```python
df1=df.drop(["ID Tag","Model","Department"],axis=1)
```

```python
from sklearn.cluster import KMeans
km=KMeans(n_clusters=2,init='k-means++',n_init=10)
km.fit(df1)
```

```python
x=km.fit_predict(df1)
print(x)
```
